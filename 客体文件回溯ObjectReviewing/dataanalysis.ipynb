```python
import dataPreprocessing as dp
import pandas as pd
import warnings
import os
from openpyxl import load_workbook
warnings.filterwarnings("ignore")
```

```python
calculateAccuracy = dp.CalculateAccuracy(repeated_times_col='RepeatedTimes')
dataPackage = dp.DataPackage(dataFolderName='rawdata')
dataPackage.readData(
    needed_num_row=192, gender_col='SubSex',
    calculateAccuracy=calculateAccuracy)
```

```python
dataMerged = dataPackage.readData(
    needed_num_row=192, gender_col='SubSex', maxNum=30, 
    accuracy_threshold=0.8,calculateAccuracy=calculateAccuracy
    )

mergedFilePath = 'dataMerged'
if not os.path.exists(mergedFilePath):
    os.makedirs(mergedFilePath)

dataMerged.to_excel(os.path.join(mergedFilePath,'dataMerged.xlsx'), index=False)
dataGenerate = dp.DataGenerate(dataMerged)
dataGenerate(verticalAxis='ReactionTime',seperatedPlots='MotionDirection',seperatedLines='TargetPosition',horizontalAxis='ISSame', subNameCol='SubName',mergedFilePath=mergedFilePath)
```
