```python
import os
import pandas as pd
import numpy as py
```

```python
def deleteTargetRow(df, target):
    """
    删除 DataFrame 中包含指定字符串的行

    Args:
    df: 输入的 DataFrame
    target_string: 要查找的字符串

    Returns:
    删除指定行后的 DataFrame
    """
    mask = ~df.apply(lambda row: row.astype(str).str.contains(target))
    dfFilterd = df[mask]
    return dfFilterd
```

```python
fileDir = os.path.join(os.path.abspath('.'), '数据')
dataMerge = pd.DataFrame()
for file in os.listdir(fileDir):
    print(file)
    filePath = os.path.join(fileDir, file)
    data = pd.read_csv(filePath, sep=',', encoding='GB18030')
    dataMerge = pd.concat([dataMerge, data])
dataMerge.head()
```

```python
dataMergeFiltered = deleteTargetRow(dataMerge, target='Data Merge Time')
resultPath = os.path.join(os.path.abspath('.'), 'dataMergeFiltered.xlsx')
dataMergeFiltered.to_excel(resultPath,index=False)

```
